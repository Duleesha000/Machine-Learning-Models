Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Import The Data Set

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
y=y.reshape(len(y),1)
```

```python
y
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc_x =StandardScaler()
sc_y =StandardScaler()

x = sc_x.fit_transform(X)
y =sc_y.fit_transform(y)
```

SVR model

```python
from sklearn.svm import SVR

regressor =SVR(kernel="rbf")
regressor.fit(x,y)
```

Prediction

```python
y_pred = regressor.predict(sc_x.transform([[6.5]])).reshape(-1,1)
y_pred
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_x.transform([[6.5]])).reshape(-1,1))
y_pred
```

Plot the curve

```python
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),c="r")
plt.plot(sc_x.inverse_transform(x),sc_y.inverse_transform([regressor.predict(x)]).reshape(10,1),c="b")
plt.title("Thruth or Bluff")
plt.xlabel("Level")
plt.ylabel("Salary")
```
