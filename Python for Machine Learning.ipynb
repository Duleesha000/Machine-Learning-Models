**Classes**

```python
class Point:
  def move(self):
    print("Move")

  def draw(self):
      print("Draw")

point1 = Point()
point1.move()
point1.draw()
```

**Constructors**

```python
class Point:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def move(self):
    print("Move")

  def draw(self):
    print("Draw")

object1 = Point(10,20)
print(object1.x)
print(object1.y)
```

**Inheritance**

```python
class Animal:
  def walk(self):
    print("Walk")

class Dog(Animal):
  def bark(self):
    print("Bark")

dog1 = Dog()
dog1.bark()
dog1.walk()
```
